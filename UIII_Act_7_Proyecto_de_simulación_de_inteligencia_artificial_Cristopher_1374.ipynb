```python
print("==========================================")
print("   ASISTENTE IA DE INVESTIGACION")
print("==========================================")

continuar = "si"

while continuar == "si":

    print("\nHola, soy una IA investigadora.")
    print("Te ayudare a analizar informacion.\n")

    nombre = input("¿Cual es tu nombre?: ")
    tema = input("¿Que tema quieres investigar?: ")
    nivel = input("Nivel de investigacion (basico/intermedio/avanzado): ")

    print("\nAnalizando informacion...")

    # Condicionales
    if nivel == "basico":
        print("\nRESULTADO DE INVESTIGACION")
        print("Tema:", tema)
        print("Nivel: Basico")
        print("La IA encontro informacion sencilla y facil de entender.")
        print("Ideal para comenzar a aprender sobre el tema.")

    elif nivel == "intermedio":
        print("\nRESULTADO DE INVESTIGACION")
        print("Tema:", tema)
        print("Nivel: Intermedio")
        print("La IA encontro datos mas detallados.")
        print("Se incluyen ejemplos y conceptos importantes.")

    elif nivel == "avanzado":
        print("\nRESULTADO DE INVESTIGACION")
        print("Tema:", tema)
        print("Nivel: Avanzado")
        print("La IA realizo un analisis profundo.")
        print("La investigacion contiene teorias y datos complejos.")

    else:
        print("\nERROR")
        print("El nivel ingresado no es valido.")

    # Segunda parte interactiva
    print("\nMENU DE OPCIONES")
    print("1. Ver recomendacion")
    print("2. Ver estado de la IA")
    print("3. Salir")

    opcion = input("Selecciona una opcion: ")

    if opcion == "1":
        print("\nRECOMENDACION")

        if tema == "inteligencia artificial":
            print("Te recomendamos investigar robots y aprendizaje automatico.")

        elif tema == "python":
            print("Te recomendamos practicar condicionales y ciclos.")

        else:
            print("La IA recomienda buscar videos y libros sobre", tema)

    elif opcion == "2":
        print("\nESTADO DE LA IA")
        print("La IA investigadora funciona correctamente.")
        print("Nivel de energia: 100%")

    elif opcion == "3":
        print("\nGracias por usar la IA investigadora.")

    else:
        print("\nOpcion no valida.")

    continuar = input("\n¿Deseas realizar otra investigacion? (si/no): ")

print("\nPrograma finalizado.")
```
